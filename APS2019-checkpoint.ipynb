```python
import os
import pandas as pd
import pyproj
from pyproj import CRS, Transformer
import rasterio
from datetime import datetime
import glob
import numpy as np

# Set specific directories
os.chdir(r"C:\Users\cbeau\2019_cycyanos_NC")
archivePath = r"C:\Users\cbeau\2019_cycyanos_NC"
txtPath = r"C:\Users\cbeau\2019_cycyanos_NC\Txt"

def projLatLon(nodes, prj_new):
    if not nodes.empty:
        llCRS = CRS.from_proj4("+proj=longlat +ellps=WGS84")
        transformer = Transformer.from_crs(prj_new, llCRS, always_xy=True)
        temp_pts = nodes[['X', 'Y']].dropna()
        indices = temp_pts.index
        transformed_coords = transformer.transform(temp_pts['X'].values, temp_pts['Y'].values)
        nodes.loc[indices, 'lon'] = transformed_coords[0]
        nodes.loc[indices, 'lat'] = transformed_coords[1]
    return nodes

try:
    # Find all images with specific pattern
    imageFiles = glob.glob(os.path.join(archivePath, "*CIcyano*.tif"))
    print(f"Found {len(imageFiles)} image files to process.")

    # Process each image file
    for image_path in imageFiles:
        print(f"Processing file: {image_path}")
        filename = os.path.basename(image_path)
        parts = filename.split('.')
        datestring = f"{parts[1]}{parts[2]}"
        date = datetime.strptime(datestring, "%Y%j%H%M")
        fntxt = os.path.join(txtPath, date.strftime("%Y%m%d%H%M.txt"))

        # Open the geotiff
        with rasterio.open(image_path) as src:
            print("Opened TIFF file.")
            prj_new = CRS.from_wkt(src.crs.to_wkt())
            rc = src.read(1)
            rows, cols = np.indices(rc.shape)
            coords = np.column_stack(src.xy(rows.flatten(), cols.flatten()))
            dfrc = pd.DataFrame(coords, columns=['X', 'Y'])
            dfrc['ci'] = rc.flatten()
            dfrc = projLatLon(dfrc, prj_new)

            mask = (dfrc['ci'] > 10) & (dfrc['ci'] <= 250)
            dfrc.loc[mask, 'CI'] = np.where(dfrc.loc[mask, 'ci'] == 0, 0, 10**(3/250 * dfrc.loc[mask, 'ci'] - 4.2))
            dfrc.loc[mask, 'CellCount'] = dfrc.loc[mask, 'CI'] * 1e8

            high_ci_points = dfrc[mask][['lon', 'lat', 'CI']].to_numpy()
            np.savetxt(fntxt, high_ci_points, fmt='%12.6f %12.6f %10.6f', header='', footer='', comments='')
            print(f"Data saved to {fntxt}")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pandas as pd
import glob
from datetime import datetime
import os

# Define the path where your files are stored
path_to_files = r"C:\Users\cbeau\2019_cycyanos_NC\Txt\*.txt"

# Step 1: Collect all file paths
file_paths = glob.glob(path_to_files)

# Step 2: Read data from each file
data_frames = []
for file_path in file_paths:
    # Check if the file is empty
    if os.stat(file_path).st_size == 0:
        print(f"Skipping empty file: {file_path}")
        continue

    try:
        # Read the file into a DataFrame, assuming no headers and comma as the delimiter
        temp_df = pd.read_csv(file_path, names=['longitude', 'latitude', 'CI'], header=None, delimiter=',')
        
        # Handle files with incomplete data
        if temp_df.isnull().values.any():
            print(f"Missing data in file: {file_path}")
            temp_df.dropna(inplace=True)  # or use `temp_df.fillna(value)`

        # Extract date from filename (adjusting path splitting for Windows)
        filename = file_path.split('\\')[-1].replace('.txt', '')
        date = datetime.strptime(filename, '%Y%m%d%H%M')
        temp_df['date'] = date
        
        # Append the DataFrame to the list
        data_frames.append(temp_df)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# Step 3: Combine all DataFrames into a single DataFrame
if data_frames:
    data = pd.concat(data_frames)
    # Step 4: Sort the DataFrame by date
    data.sort_values('date', inplace=True)
    # Step 5: Reset index in the combined DataFrame
    data.reset_index(drop=True, inplace=True)
    print(data.head())
else:
    print("No data available to process.")
```

```python
import pandas as pd
import glob
from datetime import datetime
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.animation import FuncAnimation
import numpy as np

# Define the path where your files are stored
path_to_files = r"C:\Users\cbeau\2019_cycyanos_NC\Txt\*.txt"

# Step 1: Collect all file paths
file_paths = glob.glob(path_to_files)

# Step 2: Read data from each file
data_frames = []
for file_path in file_paths:
    # Read the file into a DataFrame, assuming no headers and comma as the delimiter
    temp_df = pd.read_csv(file_path, names=['longitude', 'latitude', 'CI'], header=None, delimiter=',')

    # Extract date from filename (adjusting path splitting for Windows)
    filename = file_path.split('\\')[-1].replace('.txt', '')
    
    # Convert string to datetime object (adjust format if different)
    date = datetime.strptime(filename, '%Y%m%d%H%M')
    
    # Add a date column to the DataFrame
    temp_df['date'] = date
    
    # Append the DataFrame to the list
    data_frames.append(temp_df)

# Step 3: Combine all DataFrames into a single DataFrame
data = pd.concat(data_frames)

# Step 4: Sort the DataFrame by date
data.sort_values('date', inplace=True)

# Step 5: Reset index in the combined DataFrame
data.reset_index(drop=True, inplace=True)

print(data.dtypes)
print(data.head())

# Set up the base map
fig, ax = plt.subplots(figsize=(10, 8))
m = Basemap(projection='merc', llcrnrlat=data['latitude'].min(), urcrnrlat=data['latitude'].max(),
            llcrnrlon=data['longitude'].min(), urcrnrlon=data['longitude'].max(), lat_ts=20, resolution='i', ax=ax)

# Convert geographic coordinates to map projection coordinates
data['x'], data['y'] = m(data['longitude'].values, data['latitude'].values)

# Initialize scatter plot
scatter = m.scatter([], [], c=[], s=100, cmap='viridis', alpha=0.5, edgecolors='none', vmin=data['CI'].min(), vmax=data['CI'].max())
plt.colorbar(scatter, label='CI Value')
dates = data['date'].unique()

# Animation update function
def update(frame):
    current_date = dates[frame]
    frame_data = data[data['date'] == current_date]
    scatter.set_offsets(np.c_[frame_data['x'], frame_data['y']])
    scatter.set_array(frame_data['CI'])
    ax.set_title(f"CI Values on {current_date.strftime('%Y-%m-%d')}")
    return scatter,

# Create animation
ani = FuncAnimation(fig, update, frames=len(dates), blit=True, repeat=False)

# Save or show the animation
plt.show()
# Uncomment the line below to save the animation as an MP4 file
# ani.save('CI_values_time_series.mp4', writer='ffmpeg', fps=1)
```
