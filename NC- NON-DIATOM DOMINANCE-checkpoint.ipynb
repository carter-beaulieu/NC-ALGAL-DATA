```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load data
path = r"C:\Users\cbeau\Downloads\UNCW\Albermarle Sound Algae\NC estuarine river data 2000-2023 (1).xlsx"
ALG = pd.read_excel(path)

# Preprocess the data
ALG['Date'] = pd.to_datetime(ALG['Date'], format='%m/%d/%y')
ALG['Waterbody'] = ALG['Waterbody'].str.lower()
ALG['Year'] = ALG['Date'].dt.year
ALG['Month'] = ALG['Date'].dt.month

# Initialize LabelEncoder and encode AlgalGroup
le = LabelEncoder()
ALG['AlgalGroupEncoded'] = le.fit_transform(ALG['AlgalGroup'])
# Create a mapping for algal group names
algal_group_names = {index: label for index, label in enumerate(le.classes_)}
# Reverse mapping to get original algal group names
ALG['AlgalGroupName'] = ALG['AlgalGroupEncoded'].map(algal_group_names)

# Filter for the specified waterbodies
waterbodies_of_interest = ['neuse river', 'pamlico river', 'chowan river']
filtered_data = ALG[ALG['Waterbody'].isin(waterbodies_of_interest)]

# Determine the dominant algal group for each month and year
dominant_algal_groups = filtered_data.groupby(['Waterbody', 'Year', 'Month'], as_index=False).apply(lambda x: x.loc[x['Biovolume (mm3/m3)'].idxmax()])

# Check if the dominant algal group is Bacillariophyceae
dominant_algal_groups['IsBacillariophyceae'] = dominant_algal_groups['AlgalGroupName'].apply(lambda x: 'Yes' if x == 'Bacillariophyceae' else 'No')

# Print results
for waterbody in waterbodies_of_interest:
    print(f"Waterbody: {waterbody.capitalize()}")
    wb_data = dominant_algal_groups[dominant_algal_groups['Waterbody'] == waterbody]
    for year in sorted(wb_data['Year'].unique()):
        print(f"  Year: {year}")
        year_data = wb_data[wb_data['Year'] == year]
        for month in sorted(year_data['Month'].unique()):
            month_data = year_data[year_data['Month'] == month]
            is_bacillariophyceae = month_data.iloc[0]['IsBacillariophyceae']
            month_name = pd.to_datetime(f"{month}", format='%m').strftime('%B')
            print(f"    {month_name}: {is_bacillariophyceae}")
    print("\n")
```
